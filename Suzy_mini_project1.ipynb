The binomial distribution was initially introduced by Swiss mathematician Jacob Bernoulli in 1713. He developed the concept while studying the problem of coin tosses, where he was interested in determining the probability of getting a particular number of heads or tails in a series of flips. Bernoulli's work was then built upon by his nephew, Daniel Bernoulli, and French mathematician Pierre-Simon Laplace, who developed the modern formula for calculating the probability mass function of the binomial distribution.

The binomial distribution has become a fundamental concept in probability theory and statistics and is widely used in various fields, including biology, psychology, and finance. 

TL;DR - The founding teams were trying to create a mathematical model to describe the probability of success or failure in a series of independent events, such as coin tosses, which have only two possible outcomes. The binomial distribution is now recognized as one of the most significant and widely used probability distributions in statistics.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from numpy.random import RandomState
import scipy.stats
```

```python
# Define the parameters to be used for the binomial distribution
# Number of trials
n = 10

# Probability of success
p = 0.5
```

```python
# Call random dataset with 1000 random numbers
data = np.random.binomial(n, p, size=1000)
```

```python
# Create bins for the histogram (and shift the position closer to the centre for better readability)
bins = np.arange(0, n+2) - 0.5
```

```python
# Create the histogram with the colour pink for its bars, include a title, and labels for the axes
plt.hist(data, bins=bins, color='#ffb6c1')

plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title("Suzy's Random Binomial Distribution")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for binomial distribution
n = 10
p = 0.5

# Generate random data for binomial distribution
data = np.random.binomial(n, p, size=1000)

# Plot the final binomial distribution
plt.hist(data, bins=np.arange(0, n+2), align='left', density=True, 
         color='skyblue', edgecolor='white', alpha=0.8)
plt.title("Suzy's Binomial distribution ver.3: n=%i p=%.2f" %(n,p), fontsize=15)
plt.xlabel('Range')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define parameters for the ULTIMATE BANG-ON binomial distribution
n = 20
p = 0.7

# Generate random data for binomial distribution
binom = stats.binom(n, p)
data = binom.rvs(size=1000)

# Calculate binomial PMF
k = np.arange(0, n+1)
binomial = binom.pmf(k)

# Plot binomial distribution
plt.plot(k, binomial, 'bo', ms=8, label='Theoretical PMF')
plt.hist(data, bins=np.arange(0, n+2), align='left', density=True, 
         color='skyblue', edgecolor='white', alpha=0.8, label='Simulated Data')
plt.title("Suzy's Binomial distribution ver.4: n=%i p=%.2f" %(n,p), fontsize=15)
plt.xlabel('Range')
plt.ylabel('Probability')
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define parameters for the binomial distribution
n = 50 # Number of trials
p = 0.3 # Probability of success

# Generate random data for binomial distribution
binom = stats.binom(n, p)
data = binom.rvs(size=1000)

# Calculate binomial PMF
k = np.arange(0, n+1)
binomial = binom.pmf(k)

# Plot binomial distribution
plt.plot(k, binomial, 'o-', ms=8, color='#0077be', label='Theoretical PMF')
plt.hist(data, bins=np.arange(0, n+2), align='left', density=True, 
         color='skyblue', edgecolor='white', alpha=0.8, label='Simulated Data')
plt.title("Suzy's Binomial distribution ver.5: n=%i p=%.2f" %(n,p), fontsize=15)
plt.xlabel('Number of property reports sold')
plt.ylabel('Probability')
plt.legend(loc='best')
plt.show()
```

```python
observed = np.histogram(data, bins=np.arange(0, n+2))[0]
expected = binomial * len(data)

from scipy.stats import chisquare
chisq, p = chisquare(observed, expected)

print("Chi-squared statistic:", chisq)
print("p-value:", p)
```
